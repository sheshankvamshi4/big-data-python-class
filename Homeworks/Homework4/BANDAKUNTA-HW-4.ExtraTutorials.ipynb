# BAR CHARTS

```python
from matplotlib import pyplot as plt 
x = [5,8,10] 
y = [12,16,6]  

x2 = [6,9,11] 
y2 = [6,15,7] 

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(131)
plt.title('Bar graph')
ax1 = plt.bar(x, y2, color = 'k', align = 'center')
plt.ylabel('Y axis') 
plt.xlabel('X axis') 

ax2 = fig.add_subplot(132)
ax2 = plt.bar(x2, y, color = 'g', align = 'center') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

ax2 = fig.add_subplot(133)
ax2 = plt.bar(x2, y, color = 'b', align = 'center') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

What I Did and learnt?

I took random lists for x,y,x2,y2 and adjusted my figure size using figure() function. Then I created subplots in one row and three columns to place my three figures. Then I used bar() function for plotting the values and getting the shape as bars. There is no reason for showing three bar graphs except for showing subplots and colors.

# PIE CHART

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0.2)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

What I Did and learnt?

Demo of a basic pie chart plus a few additional features.

In addition to the basic pie chart, this demo shows a few optional features:

-slice labels: The default Labels= given listname (labels in my code) in pie() function.

-auto-labeling the percentage : The parameter autopct='%1.1f%%' gives auto labelling of percentage.

-offsetting a slice with "explode": we can use parameter explode and slice a particular item by assigning value as shown in the code.

-drop-shadow: when it is given true, It gives shadow at the eges of the pie chart.

-custom start angle: The default startangle is 0, which would start the "Frogs" slice on the
positive x-axis. This example sets startangle = 90 such that everything is
rotated counter-clockwise by 90 degrees, and the frog slice starts on the
positive y-axis.
