```python
#generate x using equally spaced number from 0 to 2*pi
x=np.linspace(0,2*np.pi,10000)
print(x)
#set the size of the plots
fig = plt.figure(figsize=(6,3))

#create subplots in the figure
#first subplot for sin of x
ax1 = fig.add_subplot(121)
plt.plot(x,np.sin(x))
plt.title("Sine of x")
#second subplot for tan of x
ax2 = fig.add_subplot(122)
plt.plot(x,np.tan(x))
plt.title("tangent of x")
```

Create a 2d histogram of 100000 points

Use the LogNorm color settings for hist2d.
And show the colorbar as well.

<img src="./images/2d_histogram.png" width="300" height="200">


```python
from matplotlib.colors import LogNorm
import matplotlib.pyplot as plt
import numpy as np


x = np.random.randn(100000)
#print(x)
y = np.random.geometric(p=0.01, size= 100000)
#print(y)

class color:
   BOLD = '\033[1m'
   END = '\033[0m'
print color.BOLD + '2d Color Histogram' + color.END

plt.hist2d(x, y, bins=50, norm=LogNorm())
plt.colorbar()
plt.show()
```
